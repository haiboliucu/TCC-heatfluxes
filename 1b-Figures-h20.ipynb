```python
#import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import xarray as xr
#import pandas as pd
#import seaborn as sb
```

```python
# First make figures of h20 - INC version
dsc  = xr.open_dataset('data/h20clim.nc')
dst  = xr.open_dataset('data/h20trend.nc')
dsc_subset = dsc.sel(Y=slice(-20,20))
dst_subset = dst.sel(Y=slice(-20,20))
#h20.plot(levels=8)
plt.figure(figsize=(8,5))

plt.subplot(211)
dp = dsc_subset.h20.plot.contourf(levels=21,cmap="jet",vmin=40,vmax=240,add_colorbar=0)
plt.title(r'INC $h_{20}$ climo')
plt.ylabel('latitude');plt.xlabel('longitude')
cbar = plt.colorbar(dp)
cbar.ax.set_ylabel(r'$h_{20}(m)$', rotation=90) 
plt.subplot(212)
dp=dst_subset.h20.plot.contourf(levels=21,cmap="RdBu_r",vmin=-30,vmax=30,add_colorbar=0)
plt.title(r'INC $h_{20}$ trend')
plt.ylabel('latitude');plt.xlabel('longitude')
cbar = plt.colorbar(dp)
cbar.ax.set_ylabel(r'$h_{20}(m)$', rotation=90) 

plt.tight_layout()
#plt.savefig('h20-INC.eps', format='eps', dpi=1000)
plt.plot()
```

```python
# Then make figures of h20 - ORAS4 version
doc  = xr.open_dataset('h20clim-ORA.nc')
dot  = xr.open_dataset('h20trend-ORA.nc')

doc_subset = doc.sel(Y=slice(-20,20))
dot_subset = dot.sel(Y=slice(-20,20))
h20c = doc_subset.h20 + dsc_subset.h20*0
h20t = dot_subset.h20 + dsc_subset.h20*0

plt.figure(figsize=(8,5))
plt.subplot(211)
dp=h20c.plot.contourf(levels=21,cmap="jet",vmin=40,vmax=240,add_colorbar=0)
plt.title(r'ORAs4 $h_{20}$ climo')
plt.ylabel('latitude');plt.xlabel('longitude')
cbar = plt.colorbar(dp)
cbar.ax.set_ylabel(r'$h_{20}(m)$', rotation=90) 
plt.subplot(212)
dp=h20t.plot.contourf(levels=21,cmap="RdBu_r",vmin=-30,vmax=30,add_colorbar=0)
plt.title(r'ORAs4 $h_{20}$ trend')
plt.ylabel('latitude');plt.xlabel('longitude')
cbar = plt.colorbar(dp)
cbar.ax.set_ylabel(r'$h_{20}(m)$', rotation=90) 

plt.tight_layout()
plt.savefig('h20-ORAs4.eps', format='eps', dpi=1000)
```
